```python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(1210)
data_points = np.random.uniform(100, 250, size=(100, 2))

df = pd.DataFrame(data_points, columns=['x', 'y'])

df.to_csv('random_dataset.csv', index=False)

fig = px.scatter(df, x='x', y='y', title="Random Veri Seti")
fig.update_layout(xaxis_title='x', yaxis_title='y')
fig.show()
```
