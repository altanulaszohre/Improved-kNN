```python
def calculate_center_and_count(input_csv, output_csv):
    df = pd.read_csv(input_csv)

    results = []
    for target in df['cluster'].unique():
        subset = df[df['cluster'] == target]

        center_x = subset['x'].mean()
        center_y = subset['y'].mean()

        distances = np.linalg.norm(subset[['x', 'y']].values - np.array([center_x, center_y]), axis=1)
        mean_distance = distances.mean()

        count = len(subset)

        results.append({
            'cluster': target,
            'center_x': center_x,
            'center_y': center_y,
            'mean_distance': mean_distance,
            'count': count,

        })

    result_df = pd.DataFrame(results)

    result_df.to_csv(output_csv, index=False)

input_csv = 'train_set.csv'
output_csv = 'train_set_calculate.csv'
calculate_center_and_count(input_csv, output_csv)
```

```python
def box_csv_files(file1, output_file, on_column='cluster'):
    df1 = pd.read_csv(file1)
    def calculate_bounding_box(df):
        grouped = df.groupby('cluster')
        results = []
        for target, group in grouped:
            min_x = group['x'].min()
            max_x = group['x'].max()
            min_y = group['y'].min()
            max_y = group['y'].max()
            center_x = (min_x + max_x) / 2
            center_y = (min_y + max_y) / 2
            mean_tsne_distance = ((max_x - min_x) ** 2 + (max_y - min_y) ** 2) ** 0.5
            area = (max_x - min_x) * (max_y - min_y)
            results.append({
                'cluster': target,
                'area': area
            })
        return pd.DataFrame(results)

    df_with_bbox = calculate_bounding_box(df1)

    df_with_bbox.to_csv(output_file, index=False)
    print(f"Done!! output: {output_file}")

file1 = 'train_set.csv'

output_file = 'train_set_box.csv'

box_csv_files(file1, output_file)
```

```python
def merge_csv_files(file1, file2, file3, output_file, on_column='cluster'):
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2)
    df3 = pd.read_csv(file3)
    merged_df1_2 = pd.merge(df1, df2, on=on_column, how='inner')
    merged_df = pd.merge(merged_df1_2, df3, on=on_column, how='inner')
    merged_df.to_csv(output_file, index=False)
    print(f"Done!! output: {output_file}")

file1 = 'train_set.csv'
file2 = 'train_set_calculate.csv'
file3 = 'train_set_box.csv'
output_file = 'merged_file.csv'

merge_csv_files(file1, file2, file3, output_file)
```
